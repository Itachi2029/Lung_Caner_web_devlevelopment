```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import seaborn
 
```

```python
dt=pd.read_csv(r"C:\Users\saura\Documents\kaggle\Cancer\survey lung cancer.csv")
```

```python
dt=dt.drop_duplicates()
dt.shape
dt['ALCOHOL CONSUMING'] = pd.to_numeric(dt['ALCOHOL CONSUMING'])
dt
```

```python
dt.head()
```

```python
dummy=pd.get_dummies(dt.GENDER)
merged=pd.concat([dt,dummy],axis='columns')
dt=merged.drop(['GENDER','F'],axis='columns')
dt
```

```python
dummy=pd.get_dummies(dt.LUNG_CANCER)
merged=pd.concat([dt,dummy],axis='columns')
dt=merged.drop(['LUNG_CANCER','NO'],axis='columns')
dt
```

```python
correlation_matrix = dt.corr()
correlation_matrix
```

```python
x=dt.drop(['YES'],axis=1)
y=dt['YES']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
clfs = RandomForestClassifier(max_depth=2, random_state=0)
clfs=clfs.fit(x_train,y_train)
y2=clfs.predict(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors = 2)
clf=clf.fit(x_train,y_train)
```

```python
y1=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y1,y_test)
```

```python
accuracy_score(y2,y_test)
```
